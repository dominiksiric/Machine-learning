```python
import matplotlib.pyplot as plt
```

```python
data = [164, 718, 6885, 6411, 7044, 7168, 7820]
labels = ['1', '2', '3', '4', '5', '6', '7']
```

```python
# Iterate over the data and labels to create line charts
for i in range(1, len(data) + 1):
    subset_data = data[:i]
    subset_labels = labels[:i]
    
    plt.plot(subset_data)
    plt.xticks(range(i), subset_labels)  # Set x-axis labels
    plt.xlabel('Submissions')
    plt.ylabel('Money saved')
    plt.title('Money saved by submission')
    
    plt.ylim(0, 8000)  # Set the y-axis range

    plt.show()  # Display the line chart
```
