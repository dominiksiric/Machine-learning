```python
from sklearn.metrics import *
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import numpy as np
import warnings
```

```python
l1 = []
duration = []
```

```python
first = "comcuc"
arrs = glob.glob(os.path.join(first, '*.labels.npy'))
print(len(arrs))
tot_sum = 0
for i in arrs:
    dt = np.load(i)
    tot_sum += np.sum(dt[:, 0] != 0)
l1.append(tot_sum)
duration.append(round(tot_sum*0.2, 1))
print(tot_sum)
print(duration)
```
