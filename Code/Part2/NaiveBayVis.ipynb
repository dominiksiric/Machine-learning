```python
import matplotlib.pyplot as plt
import numpy as np

# Results data
labels = [
    '5-fold CV (Default)', '10-fold CV (Default)', '5-fold CV (var_smoothing=0.1)',
    '5-fold CV (var_smoothing=10)', '5-fold CV (var_smoothing=1e-12)', '5-fold CV (var_smoothing=1e-20)'
]

accuracies = [
    0.7107951426507764, 0.7105618436406067, 0.7446457899269383,
    0.7113285954385088, 0.7107951426507764, 0.7107951426507764
]

f1_scores = [
    0.4722152114629634, 0.47227877261289486, 0.33090576838112157,
    0.11876147609480052, 0.4722152114629634, 0.4722152114629634
]

training_accuracies = [
    0.7120952052820544, 0.7119340889714693, 0.7449624193559587,
    0.7113285267227789, 0.7120952052820544, 0.7120952052820544
]

training_f1_scores = [
    0.4764449697062701, 0.4761454208268555, 0.33229911821546304,
    0.1187619023109769, 0.4764449697062701, 0.4764449697062701
]

# Plotting
x = np.arange(len(labels))
width = 0.2

fig, ax = plt.subplots(figsize=(12, 6))

rects1 = ax.bar(x - width, accuracies, width, label='Accuracy')
rects2 = ax.bar(x, f1_scores, width, label='F1 Score')
rects3 = ax.bar(x + width, training_accuracies, width, label='Training Accuracy')
rects4 = ax.bar(x + (2 * width), training_f1_scores, width, label='Training F1 Score')

ax.set_ylabel('Scores')
ax.set_title('Naïve Bayes (Gaussian) Results')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
# Add data labels
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

add_labels(rects1)
add_labels(rects2)
add_labels(rects3)
add_labels(rects4)

plt.tight_layout()
plt.show()
```
