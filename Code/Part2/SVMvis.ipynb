```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go

# Bar chart example using Matplotlib
x = [
    '5-fold cross-validation with default parameters',
    '10-fold cross-validation with default parameters',
    '5-fold cross-validation with C=0.1',
    '5-fold cross-validation with C=5',
    '5-fold cross-validation with kernel="poly"'
]
accuracy = [
    0.9102485443231381, 0.9113652219814414, 0.8594143261938495,
    0.919198667944551, 0.8853481248437369
]

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(x, accuracy)
ax.set_title('Accuracy of SVM Model')
ax.set_xlabel('Cross Validation')
ax.set_ylabel('Accuracy')

plt.xticks(rotation=45)
plt.subplots_adjust(top=0.9, bottom=0.2)  # Manually adjust the margins

plt.show()

# Histogram example using Plotly
hist_data = [accuracy]
group_labels = ['Accuracy']

fig = go.Figure()
for i, data in enumerate(hist_data):
    fig.add_trace(go.Histogram(x=data, name=group_labels[i]))

fig.update_layout(
    title='Accuracy Distribution of SVM Model',
    xaxis_title='Accuracy',
    yaxis_title='Count'
)

fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Results data
labels = [
    'Default (5-fold CV)', 'Default (10-fold CV)', 'C=0.1 (5-fold CV)',
    'C=5 (5-fold CV)', 'kernel="poly" (5-fold CV)'
]

accuracies = [
    0.9102485443231381, 0.9113652219814414, 0.8594143261938495,
    0.919198667944551, 0.8853481248437369
]

f1_scores = [
    0.8169193068243954, 0.8219798269285029, 0.6569836250455194,
    0.8436770480927127, 0.7555651529804173
]

training_accuracies = [
    0.9520283689937985, 0.9516566014939919, 0.8685436461349889,
    0.987770629336028, 0.9306530117814955
]

training_f1_scores = [
    0.9069535775539965, 0.9075479215337869, 0.6828261941251613,
    0.9777224255887262, 0.8590170320232481
]

# Plotting
x = np.arange(len(labels))
width = 0.2

fig, ax = plt.subplots(figsize=(14, 6))

rects1 = ax.bar(x - width, accuracies, width, label='Accuracy')
rects2 = ax.bar(x, f1_scores, width, label='F1 Score')
rects3 = ax.bar(x + width, training_accuracies, width, label='Training Accuracy')
rects4 = ax.bar(x + (2 * width), training_f1_scores, width, label='Training F1 Score')

ax.set_ylabel('Scores')
ax.set_title('SVM Results')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
# Add data labels
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

add_labels(rects1)
add_labels(rects2)
add_labels(rects3)
add_labels(rects4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Results data
labels = [
    'Default (5-fold CV)', 'Default (10-fold CV)', 'C=0.1 (5-fold CV)',
    'C=5 (5-fold CV)', 'kernel="poly" (5-fold CV)'
]

accuracies = [
    0.9102485443231381, 0.9113652219814414, 0.8594143261938495,
    0.919198667944551, 0.8853481248437369
]

f1_scores = [
    0.8169193068243954, 0.8219798269285029, 0.6569836250455194,
    0.8436770480927127, 0.7555651529804173
]

training_accuracies = [
    0.9520283689937985, 0.9516566014939919, 0.8685436461349889,
    0.987770629336028, 0.9306530117814955
]

training_f1_scores = [
    0.9069535775539965, 0.9075479215337869, 0.6828261941251613,
    0.9777224255887262, 0.8590170320232481
]

# Plotting Default (2 folds)
x_default = np.arange(2)
width = 0.2

fig, ax = plt.subplots(figsize=(8, 6))
rects1 = ax.bar(x_default - width, accuracies[:2], width, label='Accuracy')
rects2 = ax.bar(x_default, f1_scores[:2], width, label='F1 Score')
rects3 = ax.bar(x_default + width, training_accuracies[:2], width, label='Training Accuracy')
rects4 = ax.bar(x_default + (2 * width), training_f1_scores[:2], width, label='Training F1 Score')

ax.set_ylabel('Scores')
ax.set_title('Default SVM Results (2 folds)')
ax.set_xticks(x_default)
ax.set_xticklabels(labels[:2])
ax.legend()
plt.tight_layout()
plt.show()

# Plotting C (2 folds)
x_c = np.arange(2)

fig, ax = plt.subplots(figsize=(8, 6))
rects1 = ax.bar(x_c - width, accuracies[2:4], width, label='Accuracy')
rects2 = ax.bar(x_c, f1_scores[2:4], width, label='F1 Score')
rects3 = ax.bar(x_c + width, training_accuracies[2:4], width, label='Training Accuracy')
rects4 = ax.bar(x_c + (2 * width), training_f1_scores[2:4], width, label='Training F1 Score')

ax.set_ylabel('Scores')
ax.set_title('C SVM Results (2 folds)')
ax.set_xticks(x_c)
ax.set_xticklabels(labels[2:4])
ax.legend()
plt.tight_layout()
plt.show()

# Plotting kernel
x_kernel = np.arange(1)

fig, ax = plt.subplots(figsize=(8, 6))
rects1 = ax.bar(x_kernel - width, accuracies[4:], width, label='Accuracy')
rects2 = ax.bar(x_kernel, f1_scores[4:], width, label='F1 Score')
rects3 = ax.bar(x_kernel + width, training_accuracies[4:], width, label='Training Accuracy')
rects4 = ax.bar(x_kernel + (2 * width), training_f1_scores[4:], width, label='Training F1 Score')

ax.set_ylabel('Scores')
ax.set_title('Kernel SVM Results')
ax.set_xticks(x_kernel)
ax.set_xticklabels(labels[4:])
ax.legend()
plt.tight_layout()
plt.show()
```
